```python
import pandas as pd
import matplotlib.pyplot as plt 

# Load data CSV (with shuffle)
df = pd.read_csv("dataexcel-shuffle.csv", sep=";")
# Load data CSV (no shuffle)
# df = pd.read_csv("dataexcel-noshuffle.csv", sep=";")
train_acc = df["train_acc"].values
val_acc   = df["val_acc"].values
train_loss= df["train_loss"].values
val_loss  = df["val_loss"].values
```

```python
# Plot training and validation accuracy
x = [1, 2, 3, 4, 5, 6, 7] 

plt.plot(x, train_acc, marker='o', linestyle='-', color='b', label='Training Accuracy')
plt.plot(x, val_acc, marker='o', linestyle='-', color='r', label='Validation Accuracy')

plt.title('Training vs Validation Accuracy per Epoch (Excel Version)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.xticks(x)  
plt.ylim(0, 110)          
plt.legend()
plt.grid(True)
plt.show()

# Plot training and validation loss
plt.plot(x, train_loss, marker='o', linestyle='-', color='b', label='Training Loss')
plt.plot(x, val_loss, marker='o', linestyle='-', color='r', label='Validation Loss')

plt.title('Training vs Validation Loss per Epoch (Excel Version)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xticks(x)  
plt.legend()
plt.grid(True)
plt.show()
```
